```python
import json
import pandas as pd

# Load JSON from file
with open(r"drive/MyDrive/user-wallet-transactions.json", 'r') as f:
    data = json.load(f)

cleaned_data = []

for item in data:
    item['_id'] = item['_id']['$oid']

    action_data = item.pop('actionData', {})
    for key, value in action_data.items():
        item[f'actionData_{key}'] = value

    cleaned_data.append(item)

# Convert to DataFrame
df = pd.DataFrame(cleaned_data)
```

```python
from pandas import json_normalize

df = pd.json_normalize(data)
```

```python
df['actionData_amount'] = pd.to_numeric(df['actionData_amount'], errors='coerce')
df['actionData_assetPriceUSD'] = pd.to_numeric(df['actionData_assetPriceUSD'], errors='coerce')

# Create a new column: amount in USD
df['usd_amount'] = df['actionData_amount'] * df['actionData_assetPriceUSD']
```

```python
# Group by wallet and aggregate features
def compute_wallet_features(df):
    grouped = df.groupby('userWallet').agg(
        txn_count=('action', 'count'),
        num_deposits=('action', lambda x: (x == 'deposit').sum()),
        num_borrows=('action', lambda x: (x == 'borrow').sum()),
        num_repays=('action', lambda x: (x == 'repay').sum()),
        num_liquidations=('action', lambda x: (x == 'liquidationcall').sum()),
        total_deposited_usd=('usd_amount', lambda x: df.loc[x.index][df['action'] == 'deposit']['usd_amount'].sum()),
        total_borrowed_usd=('usd_amount', lambda x: df.loc[x.index][df['action'] == 'borrow']['usd_amount'].sum()),
        total_repaid_usd=('usd_amount', lambda x: df.loc[x.index][df['action'] == 'repay']['usd_amount'].sum()),
    ).reset_index()

    # Add a binary flag
    grouped['was_liquidated'] = (grouped['num_liquidations'] > 0).astype(int)

    return grouped

# Generate features
wallet_features = compute_wallet_features(df)
wallet_features.head()
```

```python
wallet_features.to_csv('wallet_features_basic.csv', index=False)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wallet_features = pd.read_csv("wallet_features_basic.csv")
wallet_features.head()
```

```python
# Normalize selected features
from sklearn.preprocessing import MinMaxScaler

score_cols = [
    'txn_count',
    'num_deposits',
    'num_borrows',
    'num_repays',
    'total_deposited_usd',
    'total_borrowed_usd',
    'total_repaid_usd',
]

X = wallet_features[score_cols].fillna(0)

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

scaled_df = pd.DataFrame(X_scaled, columns=score_cols)

# Add liquidation info
was_liquidated = wallet_features['was_liquidated'].values

# Build weighted score
score = (
    0.2 * scaled_df['total_repaid_usd'] +
    0.2 * scaled_df['num_repays'] +
    0.15 * scaled_df['total_deposited_usd'] +
    0.15 * scaled_df['txn_count'] +
    0.1 * scaled_df['num_deposits'] +
    0.1 * (1 - scaled_df['total_borrowed_usd']) +
    0.1 * (1 - was_liquidated)
)

# ðŸ”¥ This is the most important line:
wallet_features['credit_score'] = (score - score.min()) / (score.max() - score.min()) * 1000
wallet_features['credit_score'] = wallet_features['credit_score'].round(0).astype(int)
```

# Plotting the credit score distribution

```python
plt.figure(figsize=(10,5))
sns.histplot(wallet_features['credit_score'], bins=10, kde=True)
plt.title("Wallet Credit Score Distribution")
plt.xlabel("Score")
plt.ylabel("Number of Wallets")
plt.grid(True)
plt.show()
```

```python
print("Top 5 High Scores:")
print(wallet_features.sort_values(by='credit_score', ascending=False).head(5))

print("\nBottom 5 Low Scores:")
print(wallet_features.sort_values(by='credit_score', ascending=True).head(5))
```
